## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python

group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

bins = [0,9,14,19,24,29,34,39,100]

df = purchase_data[["Age","SN"]]

df1 = df.drop_duplicates()

total = df1.shape[0]

df1 = df1.drop('SN', axis=1)

df1[" "] = pd.cut(df1["Age"], bins, labels=group_names)

df1 = df1.groupby(" ").count()

df1["Percentage of Players"] =  round((df1["Age"] / total) * 100,2) 

df1 = df1.rename(columns={"Age" : "Total Count"})
#df1.columns=["Total Count","Percentage of Players"]

df1 #type(df1)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
df = {}
df1 = {}
df2 = {}
df3 = {}
df4 = {}

group_names = [" <10","10-14", "15-19","20-24","25-29","30-34","35-39","40+"]

bins = [0,9,14,19,24,29,34,39,100]

df = purchase_data[["SN","Purchase ID","Price","Age"]]
df3 = purchase_data[["SN","Age"]]

df4 = df3.copy()
df4 = df4.drop_duplicates()

df1 = df.copy()

df1[" "] = pd.cut(df1["Age"], bins, labels=group_names)
df4[" "] = pd.cut(df1["Age"], bins, labels=group_names)

df4 = df4.groupby(" ").count()

cnt = df1.groupby(" ").count() #pull purchase id

total_purchase = df1.groupby(" ").sum() #pull price

avg_purchase = round(df1.groupby(" ").mean(),2) # pull price

df2 = cnt.copy()

df2["Purchase Count"] = cnt["Purchase ID"]
df2["Average Purchase Price"] = avg_purchase["Price"]
df2["Total Purchase Value"] = total_purchase["Price"]
df2["Average Total Purchase per Person"] = round(df2["Total Purchase Value"] / df4["SN"],2)

df2["Average Purchase Price"] = df2["Average Purchase Price"].map("${:.2f}".format)
df2["Total Purchase Value"] = df2["Total Purchase Value"].map("${:.2f}".format)
df2["Average Total Purchase per Person"] = df2["Average Total Purchase per Person"].map("${:.2f}".format)

df2 = df2.drop(["Purchase ID","SN","Price","Age"], axis=1)

df2 = df2.sort_index()

#cnt
total_purchase
avg_purchase

df2

#purchase_data.columns
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
df1 = {}
df2 = {}
df3 = {}
df4 = {}
total_sum = {}

df = purchase_data[["Item ID","Item Name","Age","Price"]]
df1 = df.copy()
df1 = df1.rename(columns={"Age": "Purchase Count", "Price" : "Total Purchase Value"})
df1["Purchase Count"] = 1

df1 = df1.groupby(["Item ID","Item Name"]).sum()

df1["Item Price"] = df1["Total Purchase Value"] / df1["Purchase Count"]

df1 = df1.sort_values("Purchase Count", ascending=False)

df1 = df1[["Purchase Count","Item Price","Total Purchase Value"]]

df1["Item Price"] = df1["Item Price"].map("${:.2f}".format)
df1["Total Purchase Value"] = df1["Total Purchase Value"].map("${:.2f}".format)

df1.head()
```
